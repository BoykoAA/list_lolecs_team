```python
df = pd.read_csv('Blend_subm/1.csv')
for i in range(2, 11):
    subm = pd.read_csv('Blend_subm/' + str(i) + '.csv')
    df[i] = subm['target']
    
df.head(5)
    
```

```python
df_only_result = df.drop(columns=['card_id'])
```
